### Merge

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
})
```

```python
der = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
izq.merge(der, on = 'key')
```

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
})
der = pd.DataFrame({
    'key2': ['k0', 'k1', 'k2', 'k3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
izq.merge(der, left_on = 'key', right_on = 'key2')
```

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
})
der = pd.DataFrame({
    'key2': ['k0', 'k1', 'k2', np.nan],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
```

```python
izq.merge(der, left_on = 'key', right_on = 'key2', how = 'left')
```

```python
izq.merge(der, left_on = 'key', right_on = 'key2', how = 'inner')
```

```python
izq.merge(der, left_on = 'key', right_on = 'key2', how = 'right')
```

## Join - Index Match

```python
izq = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
}, index =  ['k0', 'k1', 'k2', 'k3'])
der = pd.DataFrame({
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}, index = ['k0', 'k1', 'k2', 'k3'],
)
```

```python
izq.join(der, how = 'left')
```

```python
izq.join(der, how = 'outer')
```

```python
izq.join(der, how = 'right')
```
